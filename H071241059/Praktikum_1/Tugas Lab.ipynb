#SOAL 1

```python
hariini = 105
kemarin = float(input('Masukkan Harga Saham Kemarin = '))

persentase = ((hariini - kemarin)/kemarin)*100
rekomedasi = (persentase > 5) * 'Beli' + (-3<= persentase <= 5 ) * 'Tahan' + (persentase < -3) * 'Jual'
print(f'Perubahan persentase harga saham: ', f"{persentase:.2f}", '%')
print(f'Rekomendasi:', rekomedasi)
```

```python
karakter = input('Masukkan Karakter: ')
kalimat = input('Masukkan Kalimat: ')

cari = karakter in kalimat
carikarakter = (f'{karakter} ditemukkan dalam "{kalimat}"') * cari + (f'{karakter} tidak ditemukan dalam "{kalimat}"') * (not cari)

print(carikarakter)
```

#SOAL 3

```python
print("Konversi detik ke jam")
detik = int(input("Masukkan jumlah detik: "))

jam = detik // 3600
menit = (detik % 3600) // 60
detik = detik % 60

print(f"{jam:02}:{menit:02}:{detik:02}")
```

#SOAL 4

```python
print('Konversi suhu dari Celcius ke Kelvin, Reamur, dan Fahrenheit')
celcius = float(input('Masukkan suhu dalam Celcius: '))

kelvin = celcius + 273
reamur = (4/5) * celcius
fahrenheit = (9/5) * celcius + 32

print(f'Hasil konversi suhu dari {celcius} derajat Celcius ke Kelvin adalah : {kelvin:.0f}'+'K')
print(f'Hasil konversi suhu dari {celcius} derajat Celcius ke Reamur adalah : {reamur:.0f}'+'R')
print(f'Hasil konversi suhu dari {celcius} derajat Celcius ke Fahrenheit adalah : {fahrenheit:.0f}'+'F')
```
